# Metodo de Euler
Se implementa el método como una función de Python `eulerVect()` que nos permitirá aproximar la solución de un problema con valores iniciales $(P.V.I)$ de la forma

$$
\left\{ \begin{array}{lcc} U^\prime (t) = F(t, U(t)), & a \leq t \leq b, \\ \\ U(a) = \alpha \end{array} \right.
$$

El problema en concreto del cual se usaran

Se importan las dependencias necesarias `math` y `matplotlib`

```python
import math
```

Se define la función que aplicara el Método de Euler a los parámetros dados

```python
def eulerVect(a: int, b: int, n: int, fun2, alpha: float):
    """Aplica el Método de Euler a los parámetros dados

    Args:
        a (int): inicio del intervalo a evaluar
        b (int): fin del intervalo a evaluar
        n (int): cantidad de aproximaciones a realizar
        fun2 (function): _description_
        alpha (float): valor del primer punto

    Returns:
        list: lista de dos vectores con las coordenadas para ejes t(tiempo) y `y1` desplazamiento respectivamente
    """
    t = []
    w = alpha
    dt = (b-a)/n

    for i in range(n):
        t.append((a+i)*dt)

    for i in range(n-1):

        col_i = [x[i] for x in w]

        f_val = fun2(t[i], col_i)

        col_next = [col_i[j] + dt * f_val[j] for j in range(len(w))]

        for row in range(len(col_i)):
            w[row].append(col_next[row])

    return [t, w]
```

```python
def fun2(t,x):
    return [x[1], -15*x[1]-200*x[0]-2*(x[0])]
```

```python
vectores = eulerVect(0,1,100,fun2,[[10],[0]])
vectores
```

```python
import matplotlib.pyplot as plt

time_axis = vectores[0]
y1 = vectores[1][0]
y2 = vectores[1][1]

# Plot data
plt.figure(figsize=(12, 9))
plt.plot(time_axis, y1, marker="o", linestyle="-", label="y1")  # Plot y1
plt.plot(time_axis, y2, marker="o", linestyle="-", label="y1")  # Plot y1

# Labels and title
plt.xlabel("Tiempo")
plt.ylabel("Valores")
plt.title("Evolucion en el tiempo de y1")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
